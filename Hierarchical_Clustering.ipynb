```python
from matplotlib import pyplot as plt
from sklearn import datasets
import numpy as np
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

### Iris Dataset

```python
iris = datasets.load_iris()
```

```python
X = iris.data
```

```python
dendrogram(linkage(X, 'complete'), orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=2, affinity='correlation', linkage='complete')
model.fit_predict(X)
```

```python
plt.scatter(x = X[:,0], y = X[:,1], c = model.labels_, cmap='rainbow' )
```
