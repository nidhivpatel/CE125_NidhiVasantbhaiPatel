## #Task 2: Apply algorithm on wine dataset - LabelEncoding of features: and Train test Division 66%-34%

```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB, MultinomialNB
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import numpy as np 
from sklearn import datasets

wine = datasets.load_wine()
```

```python
# print the names of the 13 features
print("Features: ", wine.feature_names)

# print the label type of wine(class_0, class_1, class_2)
print("Labels: ", wine.target_names)
```

```python
print("\nData shape: ",wine.data.shape)

print("\nTraget shape: ",wine.target.shape)

print("\nData type: ",type(wine.data))
newdata = wine.data[50:,:]
newtarget = wine.target[50:]

print("\nNew Data shape: ",newdata.shape)
print("\nNew Traget shape: ",newtarget.shape)
```

```python
from sklearn.model_selection import train_test_split
#split data set into train and test sets
data_train, data_test, target_train, target_test = train_test_split(newdata,
newtarget, test_size = 0.66, random_state = 125)
```

```python
import numpy as np
gnb = GaussianNB()
gnb.fit(data_train, target_train)
target_pred = gnb.predict(data_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(target_test, target_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(target_test, target_pred)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
precision = precision_score(target_test, target_pred, average='macro')
recall = recall_score(target_test, target_pred,average='macro')
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
```
