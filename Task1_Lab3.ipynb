```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB, MultinomialNB
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import numpy as np
```

```python
datasets = pd.read_csv('/content/drive/MyDrive/Machine_Learning_Labs/Lab3/Dataset3.csv') 
print(datasets)

#Use One Hot Encoder from scikit learn
onehotencoder = OneHotEncoder()
#reshape the 1-D country array to 2-D as fit_transform expects 2-D and finally fit the object 
x = onehotencoder.fit_transform(datasets.Outlook.values.reshape(-1,1)).toarray()
y = onehotencoder.fit_transform(datasets.Temp.values.reshape(-1,1)).toarray()
z = onehotencoder.fit_transform(datasets.Wind.values.reshape(-1,1)).toarray()
w = onehotencoder.fit_transform(datasets.Humidity.values.reshape(-1,1)).toarray()
```

```python
dfOneHot = pd.DataFrame(x, columns = ["Outlook_"+str(int(i)) for i in range(datasets.shape[1]-2)]) 
df = pd.concat([datasets, dfOneHot], axis=1) #column
#droping the country column 
df= df.drop(['Outlook'], axis=1) 
#printing to verify 
print(df)
datasets = df
```

```python
dfOneHot1 = pd.DataFrame(y, columns = ["Temp_"+str(int(i)) for i in range(df.shape[1]-4)]) 
df = pd.concat([df, dfOneHot1], axis=1) #column
#droping the country column 
df= df.drop(['Temp'], axis=1) 
#printing to verify 
print(df)
datasets = df
```

```python
dfOneHot2 = pd.DataFrame(z, columns = ["Humidity_"+str(int(i)) for i in range(df.shape[1]-7)]) 
df = pd.concat([df, dfOneHot2], axis=1) #column
#droping the country column 
df= df.drop(['Humidity'], axis=1) 
#printing to verify 
print(df)
datasets = df
```

```python
dfOneHot3 = pd.DataFrame(w, columns = ["Wind_"+str(int(i)) for i in range(df.shape[1]-7)]) 
df = pd.concat([df, dfOneHot3], axis=1) #column
#droping the country column 
df= df.drop(['Wind'], axis=1) 
#printing to verify 
print(df)
datasets = df
```

```python
datasets
```

```python
import seaborn as sns

corr = datasets.corr()
corr.head()
```

```python
newdata = datasets.iloc[:,1:12]
newtarget = datasets.iloc[:,0:1]
print(newdata)
print(newtarget)
```

```python
from sklearn.model_selection import train_test_split
#split data set into train and test sets
data_train, data_test, target_train, target_test = train_test_split(newdata,
newtarget, test_size = 0.85, random_state = 125)
```

```python
import numpy as np
mnb = MultinomialNB()
gnb = GaussianNB()
#Train the model using the training sets
gnb.fit(data_train, target_train)
mnb.fit(data_train, target_train)
#Predict the response for test dataset
target_pred = gnb.predict(data_test)
target_pred1 = mnb.predict(data_test)

print(target_pred)
print(target_pred1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(target_test, target_pred)
```

```python
predict = gnb.predict([[0,1,0,0,0,1,1,0,0,1,0]])
print(predict)
predict = mnb.predict([[0,1,0,0,0,1,1,0,0,1,0]])
print(predict)
```

```python
predict1 = gnb.predict([[0,0,1,0,0,1,0,1,0,0,1]])
print(predict1)
predict1 = mnb.predict([[0,0,1,0,0,1,0,1,0,0,1]])
print(predict1)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
precision = precision_score(target_test, target_pred)
precision1 = precision_score(target_test, target_pred1)
recall = recall_score(target_test, target_pred)
recall1 = recall_score(target_test, target_pred1)
print('Precision for GaussianNB: {}'.format(precision))
print('Precision for MultinomialNB: {}'.format(precision1))
print('Recall GaussianNB: {}'.format(recall))
print('Recall MultinomialNB: {}'.format(recall1))
```

```python
from sklearn import metrics
print("Accuracy for GaussianNB:",metrics.accuracy_score(target_test, target_pred))
print("Accuracy for MultinomialNB:",metrics.accuracy_score(target_test, target_pred1))
```
