Q1. Create a model that predicts crop yeilds for apples (target variable) by looking at the average temperature, rainfall and humidity (input variables or features) in different regions.

```python
import numpy as np
import torch

# input

inputs_p = np.array([[73, 67, 43],
[91, 88, 64],
[87, 134, 58],
[102, 43, 37],
[69, 96, 70]], dtype='float32')

# target

targets_p = np.array([[56],
[81],
[119],
[22],
[103]], dtype='float32')
```

```python
t_targets=torch.from_numpy(targets_p)
print(t_targets)
t_inputs=torch.from_numpy(inputs_p)
print(t_inputs)
```

```python
w = torch.randn(1, 3, requires_grad=True)
b = torch.randn(5, requires_grad=True)
print(w)
print(b)
```

```python
def model(X):
    return X @ w.t() + b

# Generating prediction and compare with targets

k=0
for x in t_inputs:
    preds = model(x)
    print("Prediction is :n",preds)
    print("Target is :g",t_targets[k])
    k+=1
```

```python
def mse_loss(predictions, targets):
    difference = predictions - targets
    return torch.sum(difference * difference)/ difference.numel()

# compute loss

k=0
for x in t_inputs:
    preds = model(x)
    print("Prediction is :n",preds)
    print("Target is :g",t_targets[k])
    print("Loss is: ",mse_loss(preds,t_targets[k]))
    k+=1
```

```python
# Computing gradient and print calculated loss 

k=0
for x in t_inputs:
      # Generate Prediction
      preds = model(x)
      loss = mse_loss(preds, t_targets[k])
      k+=1
      loss.backward()
      with torch.no_grad():
          w -= w.grad *1e-6
          b -= b.grad * 1e-6
          w.grad.zero_()
          b.grad.zero_()
print("Loss",loss)
```

```python
epochs = 100
for i in range(epochs):
    k=0
    for x in t_inputs:
        # Generate Prediction
        preds = model(x)
        loss = mse_loss(preds, t_targets[k])
        k+=1
        loss.backward()
        with torch.no_grad():
            w -= w.grad *1e-6
            b -= b.grad * 1e-6
            w.grad.zero_()
            b.grad.zero_()
    print(f"Epoch {i}/{epochs}: Loss: {loss}")
```
