# Handling Missing Value

```python
# import library
import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer 

# Data loading
        
datasets = pd.read_csv('/Exercise-CarData.csv') 
print("\nData :\n",datasets)
print("\nData statistics\n",datasets.describe())
```

```python
datasets
```

```python
X = datasets.iloc[:, :-1].values 
Y = datasets.iloc[:, -1].values
print("\n\nInput : \n", X) 
print("\n\nOutput: \n", Y) 
```

```python
datasets.dropna(axis=0,how='all',inplace=True)
print("\nNew Data :",datasets)
```

```python
updated_df = datasets;
updated_df['Age']=updated_df['Age'].fillna(updated_df['Age'].mean())
updated_df.info()
```

```python
updated_df = datasets;
updated_df['MetColor']=updated_df['MetColor'].fillna(updated_df['MetColor'].mean())
updated_df.info()
```

# Data Transformation

```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
X_new = datasets.iloc[:,1:3].values
print("\n\nX for transformation : \n", X_new)
```

```python
datasets=datasets.replace(to_replace="??",value=50000)
print(datasets)
datasets=datasets.replace(to_replace="three",value=3)
print(datasets)
```

# feature_selection_corelation

```python
import seaborn as sns
```

```python
datasets.head()
datasets.info()
```

```python
corr = datasets.corr()
corr.head()
```

```python
sns.heatmap(corr)
```

```python
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
```

```python
selected_columns = datasets.columns[j+1]
```
